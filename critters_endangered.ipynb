```python
critters_acnh['sci_name2'] = critters_acnh['sci_name'].str.extract(
    r'(?=>)(.+)(?<=</a>)')
```

```python
critters_acnh['sci_name2'] = critters_acnh['sci_name2'].fillna(critters_acnh['sci_name'])
```

```python
critters_acnh
```

```python
critters_acnh['sci_name3'] = critters_acnh['sci_name2'].str.replace(r'>',"").str.replace(r'<','').str.replace(r'/a','')
```
